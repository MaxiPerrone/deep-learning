```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
iris_dataset.target_names
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.c_[iris_dataset['data'], iris_dataset['target']], columns= iris_dataset['feature_names'] + ['target'])
df
```

```python
df_reduced = df[['petal length (cm)', 'petal width (cm)', 'target']]
df_reduced
```

```python
X = df_reduced.drop('target', axis=1)
y = df_reduced['target'].copy()
```

```python
import matplotlib.pyplot as plt

X.plot.scatter('petal length (cm)', 'petal width (cm)')
plt.show()
```

```python
from sklearn.linear_model import Perceptron

model = Perceptron(max_iter=1000, random_state=40)
model.fit(X, y)
```

**Perceptron multicapa:**

```python
from sklearn.neural_network import MLPClassifier

model = MLPClassifier()
model.fit(X, y)
```

```python
# Número de capas del perceptrón multicapa
model.n_layers_
```
